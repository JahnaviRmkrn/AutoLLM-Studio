### Importing libraries

```python
import json
import os
```

```python
# Making sure datasets folder exists
os.makedirs("../datasets", exist_ok = True)
```

```python
dataset = []
```

```python
tasks = [
"Fix leaking tap",
"Unclog kitchen drain",
"Install ceiling fan",
"Repair electrical switch",
"Clean water tank",
"Fix door hinge",
"Install RO purifier",
"Repair washing machine",
"Paint room wall",
"Install geyser",
"Fix window lock",
"Repair electrical socket",
"Install bathroom shower",
"Clean AC filter",
"Replace tube light"
]
```

```python
for i in range(8):
     for task in tasks:
          dataset.append({
               "instruction": task,
               "input": "",
               "output": f"To {task.lower()}, first ensure safety precautions are followed, gather the required tools, and complete the service step-by-step."
          })
```

```python
file_path = "../datasets/household_services_dataset.json"
```

```python
with open(file_path, "w") as f:
     json.dump(dataset, f, indent = 4)
```

```python
print(f"Dataset saved at: {file_path}")
print(f"Total sample: {len(dataset)}")
```
